# Other Charts

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
dict = {'ages' : [18,18,19,21,25,26,26,30,32,38,45,55,45,45,25,18,18]}
df = pd.DataFrame(dict)
df
```

## Cumulative Frequency Plot 

```python
plt.style.use('fivethirtyeight')
bins = [10,20,30,40,50,60]
plt.hist(df['ages'], cumulative=True,bins=bins ,edgecolor='white')
plt.title('Cumulative Frequency of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Relative Frequency Plot

```python
plt.style.use('fivethirtyeight')
bins = [10,20,30,40,50,60]
plt.hist(df['ages'], density=True, color='lightgreen',bins=bins ,edgecolor='white')
plt.title('Relative Frequency of Ages')
plt.xlabel('Age')
plt.ylabel('Relative Frequency')
plt.show()
```

## Cumulative Relative Frequency Plot

```python
plt.style.use('fivethirtyeight')
bins = [10,20,30,40,50,60]
plt.hist(df['ages'], density=True,cumulative=True, color='navy',bins=bins, edgecolor='white')
plt.title('Relative Cumulative Frequency of Ages')
plt.xlabel('Age')
plt.ylabel('Relative Frequency')
plt.show()
```
