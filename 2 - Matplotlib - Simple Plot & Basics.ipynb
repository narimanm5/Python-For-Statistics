```python
from matplotlib import pyplot as plt
import pandas as pd
```

## Creating Lists

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
# They should have the same length!
```

## Simple Plot
plt.plot()

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

## Adding Title and Labels to the Plot
plt.title() \
plt.xlabel() \
plt.ylabel()

```python
plt.plot(x,y)
plt.title('Simple Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## Legend
plt.legend()

```python
#first Way
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [20,30,40,50,60]
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['y','z'])
plt.show()
```

```python
#2nd way - Best way
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [20,30,40,50,60]
plt.plot(x,y ,label = 'y')
plt.plot(x,z ,label = 'z')
plt.legend()
plt.show()
```

## Grids and Color, Marker & Linestyle
plt.grid()

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [20,30,40,50,60]
plt.plot(x,y ,label = 'y',color = 'red',marker = 'o',linestyle='--')
plt.plot(x,z ,label = 'z',color = 'blue',marker = 'o',linestyle='-')
plt.legend()
plt.grid(True)
plt.show()
```

## PLT Styles

```python
plt.style.use('fivethirtyeight')
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [20,30,40,50,60]
plt.plot(x,y ,label = 'y')
plt.plot(x,z ,label = 'z')
plt.legend()
plt.show()
```

## Save as PNG

```python
plt.style.use('fivethirtyeight')
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [20,30,40,50,60]
plt.plot(x,y ,label = 'y')
plt.plot(x,z ,label = 'z')
plt.legend()
plt.savefig('Plot.png')
plt.show()
```

## Plotting Data Frames

```python
data = {
    'year': [2010, 2011, 2012, 2013, 2014],
    'gdp': [14.48, 15.54, 16.2, 16.69, 17.35],
    'unemployment_rate': [9.6, 9.0, 8.1, 7.4, 6.2]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['gdp'], marker='o', linestyle='-', color='b', label='GDP')
plt.plot(df['year'], df['unemployment_rate'], marker='s', linestyle='--', color='r', label='Unemployment Rate')
plt.title('GDP and Unemployment Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```
