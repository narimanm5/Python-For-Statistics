# Bar Charts
plt.bar()

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.bar(x,y)
plt.title('Bar Chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Width

```python
plt.style.use('fivethirtyeight')
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.bar(x,y ,width = 0.25)
plt.show()
```

## Multiple Bar Charts, Legend & Labels

```python
x_indexes = np.arange(len(x))
x_indexes
```

```python
plt.style.use('fivethirtyeight')
width = 0.25
x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [20,30,40,50,60]
t = [30,40,50,60,70]
plt.bar(x_indexes-width,y,label='y',width=width)
plt.bar(x_indexes,z,label='z',width=width)
plt.bar(x_indexes+width,t,label='t',width=width)
plt.legend()
plt.title('Bar Chart')
plt.xlabel('X')
plt.ylabel('Y,Z & T')
plt.show()
```

## Horizontal Bar
plt.barh()

```python
plt.style.use('fivethirtyeight')
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.barh(x,y ,label = 'y')
plt.legend()
plt.title('Bar Chart')
plt.xlabel('Y')
plt.ylabel('X')
plt.show()
#Better when you have a lot of values or strings as X
```
