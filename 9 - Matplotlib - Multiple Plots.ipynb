# Multiple Plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Data/Scatter.csv')
df
```

```python
plt.style.use('fivethirtyeight')

plt.plot(df['GDP'],df['Inflation'] ,label = 'Inflation')
plt.plot(df['GDP'],df['Unemployment'] ,label = 'Unemployment')
plt.plot(df['GDP'],df['Interest_Rate'], label = 'Interest_Rate')

plt.legend()

plt.title('Economic Indicators by GDP')
plt.xlabel('GDP')
plt.ylabel('Percentage%')

plt.tight_layout()
plt.show()
```

## Subplot Method

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(df['GDP'],df['Inflation'] ,label = 'Inflation')
ax.plot(df['GDP'],df['Unemployment'] ,label = 'Unemployment')
ax.plot(df['GDP'],df['Interest_Rate'], label = 'Interest_Rate')

ax.legend()

ax.set_title('Economic Indicators by GDP')
ax.set_xlabel('GDP')
ax.set_ylabel('Percentage%')

plt.tight_layout()
plt.show()
```

## Multiple Plots in one figure

```python
plt.style.use('fivethirtyeight')

fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(8, 6))
#sharex=True
ax1.plot(df['GDP'],df['Inflation'] ,label = 'Inflation')
ax2.plot(df['GDP'],df['Unemployment'] ,label = 'Unemployment',color='red')
ax2.plot(df['GDP'],df['Interest_Rate'], label = 'Interest_Rate',color='orange')

ax1.legend()
ax1.set_title('Economic Indicators by GDP')
ax1.set_xlabel('GDP')
ax1.set_ylabel('Percentage%')

ax2.legend()
ax2.set_title('Economic Indicators by GDP')
ax2.set_xlabel('GDP')
ax2.set_ylabel('Percentage%')

plt.tight_layout()
plt.show()
```

## Multiple Charts in one Plot

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(df['GDP'],df['Inflation'] ,label = 'Inflation',width = 10)
ax2 = ax.twinx()
ax2.plot(df['GDP'],df['Unemployment'] ,label = 'Unemployment',color='red')
ax2.plot(df['GDP'],df['Interest_Rate'], label = 'Interest_Rate',color='orange')

ax.legend()
ax.set_title('Economic Indicators by GDP')
ax.set_xlabel('GDP')
ax.set_ylabel('Percentage%')
ax.set_ylim(0,6)

ax2.legend()
ax2.set_ylim(0,6)

plt.tight_layout()
plt.show()
```
