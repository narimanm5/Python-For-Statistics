# Scatter Plots

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Data/Scatter.csv')
df.drop('Unnamed: 0',axis=1,inplace = True)
df
```

```python
gdp = df['GDP']
inflation = df['Inflation']
unemployment = df['Unemployment']
interest_rate = df['Interest_Rate']
```

## Scatter Plot & Arguments

```python
plt.style.use('fivethirtyeight')
plt.scatter(inflation,unemployment , s=100,c='green',marker='o',alpha = 0.5,edgecolor='black',linewidth=2)
plt.show()
```

## Adding Third Variable in Scatter Plots
showing the third variable with colors
 <br> color map,  <br>  and adding labels

```python
plt.style.use('fivethirtyeight')
plt.scatter(inflation,unemployment , s=500,c=interest_rate,cmap='Greens',edgecolor='black')
cbar = plt.colorbar()
cbar.set_label('GDP')
plt.show()
```

## Adding Fourth Variable in Scatter Plots
showing the fourth variable with size


```python
plt.style.use('fivethirtyeight')
plt.scatter(inflation,unemployment , s=gdp,c=interest_rate,cmap='Greens',edgecolor='black')
cbar = plt.colorbar()
cbar.set_label('GDP')
plt.show()
```
