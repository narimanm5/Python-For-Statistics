# Pie Charts
plt.pie()

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
slices = [60,40]
plt.pie(slices)
plt.title('Pie Chart')
plt.show()
```

## Labels

```python
plt.style.use('fivethirtyeight')
slices = [60,40]
labels = ['Sixty','Forty']
plt.pie(slices,labels = labels)
plt.title('Pie Chart')
plt.show()
```

## Wedge & Slices 
Edge Color - Slice Color

```python
#Changing the edge color
plt.style.use('fivethirtyeight')
slices = [60,40]
labels = ['Sixty','Forty']
plt.pie(slices,labels = labels,wedgeprops = {'edgecolor':'black'})
plt.title('Pie Chart')
plt.show()
```

```python
#Change the color of Slices
plt.style.use('fivethirtyeight')
slices = [60,40]
labels = ['Sixty','Forty']
colors = ['green','yellow']
plt.pie(slices,labels = labels,wedgeprops = {'edgecolor':'black'}, colors = colors)
plt.title('Pie Chart')
plt.show()
```

## Explode
Emphasize on one slice of the chart

```python
plt.style.use('fivethirtyeight')
slices = [60,40,20,10,5]
labels = ['Sixty','Forty','Twenty','Ten','Five']
colors =  ['#FF005C', '#6B8E23', '#800080', '#FFD700', '#4682B4']
explode = [0,0,0,0.1,0]
plt.pie(slices,labels = labels, explode = explode,colors=colors )
plt.title('Pie Chart')
plt.show()
```

## Percentages

```python
plt.style.use('fivethirtyeight')
slices = [60,40,20,10,5]
labels = ['Sixty','Forty','Twenty','Ten','Five']
plt.pie(slices,labels = labels ,autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
